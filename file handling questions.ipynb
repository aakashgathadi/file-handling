```python
#11 Create a student record system that allows adding student data (name, roll, marks) and saving the data to a file.
f = open("student.txt", "a")

while True:
    name = input("Enter Name: ")
    roll = input("Enter Roll: ")
    marks = input("Enter Marks: ")

    f.write(name + "," + roll + "," + marks + "\n")

    choice = input("Do you want to add another student? (yes/no): ")

    if choice.lower() != "yes":
        break

f.close()

print("Records Added Successfully")
```

```python
#12 Search a student by roll number.
roll = input("Enter Roll Number: ")

f = open("student.txt", "r")

for line in f:
    data = line.strip().split(",")

    if data[1] == roll:
        print("Name:", data[0])
        print("Roll:", data[1])
        print("Marks:", data[2])

f.close()
```

```python
#13  Add functionality to update student marks.
roll = input("Enter Roll Number: ")
new_marks = input("Enter New Marks: ")

f = open("student.txt", "r")
lines = f.readlines()
f.close()

f = open("student.txt", "w")

for line in lines:
    data = line.strip().split(",")

    if data[1] == roll:
        line = data[0] + "," + data[1] + "," + new_marks + "\n"

    f.write(line)

f.close()
```

```python
#14 Delete a student record by name or roll number
roll = input("Enter Roll Number: ")

f = open("student.txt", "r")
lines = f.readlines()
f.close()

f = open("student.txt", "w")

for line in lines:
    data = line.strip().split(",")

    if data[1] != roll:
        f.write(line)

f.close()

print("Record Deleted")
```

```python
#15 Show all student records sorted by marks.
f = open("student.txt", "r")

students = []

for line in f:
    data = line.strip().split(",")
    students.append(data)

f.close()

students.sort(key=lambda x: int(x[2]), reverse=True)

for s in students:
    print("Name:", s[0], "Roll:", s[1], "Marks:", s[2])
```

```python
#16 Display student records with marks greater than 75.
f = open("student.txt", "r")

for line in f:
    data = line.strip().split(",")

    if int(data[2]) > 25:
        print("Name:", data[0], "Roll:", data[1], "Marks:", data[2])

f.close()
```

```python
#17 Export top 3 performing students to a new file.
f = open("student.txt", "r")

students = []

for line in f:
    data = line.strip().split(",")
    students.append(data)

f.close()

students.sort(key=lambda x: int(x[2]), reverse=True)

f = open("top3.txt", "w")

for s in students[:3]:
    f.write(",".join(s) + "\n")

f.close()

print("Top 3 Students Exported")
```

```python
#19 Count how many students scored more than 90 and less than 35
f = open("student.txt", "r")

above90 = 0
below35 = 0

for line in f:
    data = line.strip().split(",")

    marks = int(data[2])

    if marks > 90:
        above90 += 1

    if marks < 35:
        below35 += 1

f.close()

print("Above 90:", above90)
print("Below 35:", below35)
```

```python
#20 Create a menu-driven system (Add, View, Search, Update, Delete, Exit).
while True:

    print("\n1. Add")
    print("2. View")
    print("3. Search")
    print("4. Update")
    print("5. Delete")
    print("6. Exit")

    ch = int(input("Enter Choice: "))

    if ch == 1:
        f = open("student.txt", "a")
        name = input("Name: ")
        roll = input("Roll: ")
        marks = input("Marks: ")
        f.write(name + "," + roll + "," + marks + "\n")
        f.close()

    elif ch == 2:
        f = open("student.txt", "r")
        print(f.read())
        f.close()

    elif ch == 3:
        roll = input("Enter Roll: ")
        f = open("student.txt", "r")
        for line in f:
            data = line.strip().split(",")
            if data[1] == roll:
                print(line)
        f.close()

    elif ch == 4:
        roll = input("Enter Roll: ")
        new_marks = input("New Marks: ")

        f = open("student.txt", "r")
        lines = f.readlines()
        f.close()

        f = open("student.txt", "w")

        for line in lines:
            data = line.strip().split(",")

            if data[1] == roll:
                line = data[0] + "," + data[1] + "," + new_marks + "\n"

            f.write(line)

        f.close()

    elif ch == 5:
        roll = input("Enter Roll: ")

        f = open("student.txt", "r")
        lines = f.readlines()
        f.close()

        f = open("student.txt", "w")

        for line in lines:
            data = line.strip().split(",")

            if data[1] != roll:
                f.write(line)

        f.close()

    elif ch == 6:
        print("Program Ended")
        break

    else:
        print("Invalid Choice")
```
